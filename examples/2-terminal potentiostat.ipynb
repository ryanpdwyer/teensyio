```python
t = TeensyIO(find_teensy())
```

```python
t.send_cmd('p');
t.setup_cyclic_voltamettry(V_start=2048, V_max=3196, V_min=2048, V_end=2048, N_cycles=2)
t.send_cmd('p');
```

```python
t.run_and_plot()
```

```python
t.df['y_scaled'] = (t.df.y - 2048) / 2048. * 10
```
